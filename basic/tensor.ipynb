```python
# 学习pytorch中tensor的基本使用方法
%matplotlib inline
import torch
import numpy as np
```

```python
torch.__version__
```

```python
torch.cuda.is_available()
```

```python
# 初始化tensor
# 直接从数据初始化
data = [
    [1, 2],
    [3, 4]
]
x_data = torch.tensor(data)
x_data
```

```python
# 从numpy array初始化
np_array = np.array(data)
x_np = torch.from_numpy(np_array)

print(f"Numpy np_array value: \n {np_array} \n")
print(f"Tensor x_np value: \n {x_np} \n")

# np_array * 2
np.multiply(np_array, 2, out=np_array)

print(f"Numpy np_array after * 2 operation: \n {np_array} \n")
print(f"Tensor x_np value after modifing numpy array: \n {x_np} \n")
```

```python
# 从其他tensor初始化
x_ones = torch.ones_like(x_data) # 保持x_data的属性
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # 覆盖x_data的属性
print(f"Rndom Tensor: \n {x_rand} \n")
```

```python
# shape是用来描述tensor的维度的元组
# 代表2行3列
shape = (2, 3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor} \n")
```

```python
# Tensor常见属性：shape、data type、device
tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of Tensor: {tensor.dtype}")
print(f"Device tensor is stored on : {tensor.device}")
```

```python
# Move tensor to gpu
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
```

```python
# indexing and slicing
tensor = torch.ones(4, 4)
print('First row: ', tensor[0])
print('First column: ', tensor[:,0])
# Ellipsis is used here to indicate a placeholder
# for the rest of the array dimensions not specified.
# e.g. for a 4d tensor, a[0,...,0] equals a[0,:,:,0]
# [..., -1] equals [:, -1] here
print('Last column: ', tensor[..., -1])
tensor[:, 1] = 0
print(tensor)
```

```python
# join/concatenate tensors（torch.cat or torch.stack）
t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)
```

```python
# arithmetic operations
# matrix multiplication
y1 = tensor @ tensor.T
y2 = tensor.matmul(tensor.T)

y3 = torch.rand_like(tensor)
torch.matmul(tensor, tensor.T, out=y3)

# element-wise product(dot product)
z1 = tensor * tensor
z2 = tensor.mul(tensor)

z3 = torch.rand_like(tensor)
torch.mul(tensor, tensor, out=z3)

print('y1: ', y1)
print('y2: ', y2)
print('y3: ', y3)

print('z1: ', z1)
print('z2: ', z2)
print('z3: ', z3)
```

```python
# use item() convert a single-element tensor to a python numerical value
agg = tensor.sum()
agg_item = agg.item()
print(agg_item, type(agg_item))
```

```python
# in-place operation, their use is discouraged, will lost operand
print(tensor, '\n')
tensor.add_(5)
print(tensor)
```

```python
# bridge with numpy
# share underlying memory between tensor and numpy array
# changing one will change the other
t = torch.ones(5)
print(f't: {t}')
n = t.numpy()
print(f'n: {n}') 
```

```python
t.add_(1)
print(f't: {t}')
print(f'n: {n}')
```

```python
n = np.ones(5)
t = torch.from_numpy(n)
```

```python
np.add(n, 1, out=n)
print(f't: {t}')
print(f'n: {n}')
```
